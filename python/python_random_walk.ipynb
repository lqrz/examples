## A simple random walk example: cumsum

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# config params
n_steps = 50
initial_point = 0
```

```python
walking_steps = []

for ix in range(n_steps):
#     with uniform prob, takes step 1 or -1
    previous_step = walking_steps[ix-1] if ix-1 > 0 else initial_point
    next_step = 1 if np.random.random() > .5 else -1
    walking_steps.append(previous_step + next_step)
```

```python
_, ax = plt.subplots(figsize=(15,5))
sns.pointplot(x=np.arange(n_steps), y=walking_steps, ax=ax)
```

```python
# The previous implementation is just doing a cumsum of the steps. Implemented that way
walking_steps = np.cumsum([1 if np.random.random() > .5 else -1 for _ in range(n_steps)])
```

```python
_, ax = plt.subplots(figsize=(15,5))
sns.pointplot(x=np.arange(n_steps), y=walking_steps, ax=ax)
```
