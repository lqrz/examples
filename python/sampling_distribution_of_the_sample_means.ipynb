# The sampling distribution of the sample means


```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

The original distribution is completely unrelated to the Normal distribution.

```python
probs = np.array([.3, .1, .05, .15, .4])

sb.barplot(x=np.arange(probs.shape[0]), y=probs, color='navy')

print('Distrbution mean: %f' % np.dot(np.arange(probs.shape[0]), probs))
```

When we use a low sample size, we do not get to see anything special.

```python
n_samples = 3
cnts = Counter([np.mean(np.random.choice(5, n_samples , p=probs)) for _ in range(10000)])

sb.barplot(x=list(cnts.keys()), y=list(cnts.values()), color='navy')
```

According to the Central limit theorem, when the sample size goes to infinity, the sampling distribution of the sample means is going to be Gaussian.
Even with a sample size of 10 the distribution is already Normal-like.

```python
n_samples = 10
cnts = Counter([np.mean(np.random.choice(5, n_samples , p=probs)) for _ in range(10000)])

sb.barplot(x=list(cnts.keys()), y=list(cnts.values()), color='navy')
```
